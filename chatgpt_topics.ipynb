```python
from openai import OpenAI
import pandas as pd

from dotenv import load_dotenv
import os

load_dotenv()
client = OpenAI(
    api_key=os.getenv("CHATGPT_KEY"),
)

def extract_key_topics(text):
    response = client.chat.completions.create(
        model = "gpt-4o-mini",
        messages = [{
            "role":"user",
            "content":"Summarise the key topics from the following text in 7 to 10 points: " + text
        }],
        max_completion_tokens = 150
    )

    return response.choices[0].message.content

df = pd.read_csv("data/all-responses.csv")
combined_string = df['Q23 - Do you have any ideas for new technology or examples you\'ve seen from other...']
topics = extract_key_topics(combined_string)
print(topics)
```
